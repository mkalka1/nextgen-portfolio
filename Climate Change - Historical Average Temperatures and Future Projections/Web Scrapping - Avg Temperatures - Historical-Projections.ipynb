### Load the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Import graph data in JSON format from the Request URL into the data frame

```python
df_1901_1930 = pd.read_json (r'https://climatedata.worldbank.org/climateweb/rest/v1/cru/country/tas/1901-1930/USA')
df_1931_1960 = pd.read_json (r'https://climatedata.worldbank.org/climateweb/rest/v1/cru/country/tas/1931-1960/USA')
df_1961_1990 = pd.read_json (r'https://climatedata.worldbank.org/climateweb/rest/v1/cru/country/tas/1961-1990/USA')
df_1991_2016 = pd.read_json (r'https://climatedata.worldbank.org/climateweb/rest/v1/cru/country/tas/1991-2016/USA')
df_1901_2016 = pd.read_json (r'https://climatedata.worldbank.org/climateweb/rest/v1/cru/country/tas/1901-2016/USA')
```

### Append all data frames into one single data frame
### Display Data frame - Average Monthly temperatures for all year ranges

```python
df_historical = df_1901_1930.append(df_1931_1960, ignore_index = True).append(df_1961_1990, ignore_index = True).append(df_1991_2016, ignore_index = True).append(df_1901_2016, ignore_index = True)
df_historical
```

### Delete unwanted column - AgrLevel

```python
df_historical.drop('AgrLevel', axis=1, inplace=True)
df_historical
```

### Change order of the header columns - Swap YearRang with monthVals

```python
df_historical = df_historical[['monthVals', 'YearRang','variable','Code']]
df_historical
```

### Replace Headers

```python
df_historical = df_historical.rename(columns={'monthVals': 'Average_Temperature_Deg_C'})
df_historical = df_historical.rename(columns={'YearRang': 'Year'})
df_historical = df_historical.rename(columns={'variable': 'Month'})
df_historical = df_historical.rename(columns={'Code': 'Country'})
df_historical
```

### Format data

#### Replace data under Month column with corresponding Month values

```python
 for index in range(0,len(df_historical)):
    if index == 0 or index == 12 or index == 24 or index == 36 or index == 48:
        df_historical.at[index,'Month'] = 'January'
    elif  index == 1 or index == 13 or index == 25 or index == 37 or index == 49:
        df_historical.at[index,'Month'] = 'February'
    elif  index == 2 or index == 14 or index == 26 or index == 38 or index == 50:
        df_historical.at[index,'Month'] = 'March'
    elif  index == 3 or index == 15 or index == 27 or index == 39 or index == 51:
        df_historical.at[index,'Month'] = 'April'
    elif  index == 4 or index == 16 or index == 28 or index == 40 or index == 52:
        df_historical.at[index,'Month'] = 'May'
    elif  index == 5 or index == 17 or index == 29 or index == 41 or index == 53:
        df_historical.at[index,'Month'] = 'June'
    elif  index == 6 or index == 18 or index == 30 or index == 42 or index == 54:
        df_historical.at[index,'Month'] = 'July'
    elif  index == 7 or index == 19 or index == 31 or index == 43 or index == 55:
        df_historical.at[index,'Month'] = 'August'
    elif  index == 8 or index == 20 or index == 32 or index == 44 or index == 56:
        df_historical.at[index,'Month'] = 'September'
    elif  index == 9 or index == 21 or index == 33 or index == 45 or index == 57:
        df_historical.at[index,'Month'] = 'October'
    elif  index == 10 or index == 22 or index == 34 or index == 46 or index == 58:
        df_historical.at[index,'Month'] = 'November'
    elif  index == 11 or index == 23 or index == 35 or index == 47 or index == 59:
        df_historical.at[index,'Month'] = 'December'

df_historical
```

### Check for duplicates

```python
print("Average Temperature is duplictaed - {}".format(any(df_historical.Average_Temperature_Deg_C.duplicated())))
print("Year is duplictaed - {}".format(any(df_historical.Year.duplicated())))
print("Month is duplictaed - {}".format(any(df_historical.Month.duplicated())))
print("Country is duplictaed - {}".format(any(df_historical.Country.duplicated())))
```

### Check for NaN values

```python
print("The column Average Temperature contains NaN - %r " % df_historical.Average_Temperature_Deg_C.isnull().values.any())
print("The column Year contains NaN - %r " % df_historical.Year.isnull().values.any())
print("The column Month contains NaN - %r " % df_historical.Month.isnull().values.any())
print("The column Country contains NaN - %r " % df_historical.Country.isnull().values.any())
```

### Fix inconsistent values

####  Average Temperature column has inconsistent decimal places

#### Round up Average Temperature Values to "Two" decimals

```python
df_historical["Average_Temperature_Deg_C"] = round(df_historical.Average_Temperature_Deg_C, 2)
df_historical
```

### Identify Outliers - Create a box plot

```python
plt.boxplot(df_historical.Average_Temperature_Deg_C, notch=True)
```

```python
# Find maximum average historical temperature
df_historical.Average_Temperature_Deg_C.max()
```

```python
# Find minimum average historical temperature
df_historical.Average_Temperature_Deg_C.min()
```

#### Data in this column in the interval (-6.3 Deg C, 20.46 Deg C). However, the main concentration of the data is between -0.5 Deg C to 15.5 Deg C. Anything beyond 20 Deg C and bellow -6 Deg C are outliers.

### Report the size before getting rid of the outliers

```python
df_historical.shape
```

### Get rid of the outliers

```python
df_historical_minus_outliers = df_historical[(df_historical['Average_Temperature_Deg_C'] < 20) & (df_historical['Average_Temperature_Deg_C'] > -6)]
```

### Report the size after getting rid of the outliers

```python
df_historical_minus_outliers.shape
```

### Export Dataframe into Output csv File - Historical Data

```python
df_historical_minus_outliers.to_csv('Historical_Data_Average_Temperatures_USA.csv', index=False)
```
